```python
import pandas as pd
```

```python
# Load the original CSV
df = pd.read_csv('your_file_path.csv')
```

```python
df.head()
```

```python
pivot_table = pd.pivot_table(df, 
                             values=['ewa_amount','membership_fee','membership_fee_disc','disbursement_fee','outstanding_principal','outstanding_non_principal','outstanding_non_principal_with_tax'],  # Multiple columns
                             index='reimbursement_notice_id',            # Row categories
                             # columns='reimbursement_notice_id',           # Column categories
                             aggfunc='sum')               # Aggregation function


```

```python
# Display the pivot table
print(pivot_table)
```

```python
# Save to CSV
pivot_table.to_csv('pivot_output.csv')
```
